````python
from databricks_langchain import ChatDatabricks
````

````python
# Define your schema (JSON Schema)
schema = {
    "type": "object",
    "properties": {
        "product_name": {"type": "string"},
        "category": {"type": "string"},
        "sentiment": {"type": "string"},
        "score": {"type": "number"}
    },
    "required": ["product_name", "category", "sentiment", "score"],
}

endpoint = 'databricks-claude-sonnet-4-5'

# Initialize the DatabricksChatModel with structured output
chat = ChatDatabricks(
    endpoint=endpoint,
    output_schema=schema
)

# Example retail-style prompt
messages = [
    {
        "role": "user",
        "content": (
            "Extract the following fields as JSON: product_name, category, sentiment, and score "
            "from this customer review: 'The SpaceJet wireless headphones from AudioWorld are amazing. "
            "I love the sound quality and the comfort. I give them a score of 4.8.'"
        ),
    }
]

response = chat.invoke(messages)
print("Structured Output:", response)
````

````python
import json
json.loads(response.content.replace("json\n","").replace("```",""))
````
